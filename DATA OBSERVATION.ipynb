```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data import
train_df = pd.read_csv("./Lightning7/Lighting7_TRAIN.txt", header=None)
test_df = pd.read_csv("./Lightning7/Lighting7_TEST.txt",header=None)
all_data = pd.concat([train_df, test_df]) #combining data to experiment with different train/test splits
```

```python
all_data.head()
```

```python
#showing values by class
classes = {}
for i in range(7):
    classes[i] = all_data.loc[all_data[0]==i]
```

```python
#plot some representative examples together
for c in classes:
    plt.figure(c)
    plt.title("Class %d"%c)
    plt.plot(classes[c].sample(n=1).drop(0,axis=1).transpose().values)
```

```python
#plot classes together
for i in range(10):
    plt.figure(i)
    for c in classes:
        plt.plot(classes[c].sample(n=1).drop(0,axis=1).transpose().values, label="Class %d"%c)
        plt.legend()
```
